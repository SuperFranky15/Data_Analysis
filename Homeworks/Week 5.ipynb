### 2/a)

```python
mydata1=woo.data('FERTIL2')
mydata1
```

```python
model1 = smf.ols('children~educ+age+I(age**2) ', mydata1).fit()
model1.summary()
```

From the regression estimates above, when age is fixed, the estimated effect of another year of education on fertility is -0.0906. If 100 women receive another year of education, they will expected to have 0.0906 * 100 fewer children, which is approximately 9 children.

### 2/b)

```python
model2 = smf.ols('educ~frsthalf ', mydata1).fit()
model2.summary()
```

```python
hypothesis = ['frsthalf = 0']
model2.f_test(hypothesis)
```

From the result of F-test above, we can see that F-test has p-value equal to 3.1539715657891034e-15, which is extremely small. This means F-statistic is significant in this model and it suggests that frsthalf is a strong IV.

### 2/c)

```python
mydata1["educ_hat"]=model2.fittedvalues
model3 = smf.ols('children~educ_hat+age+I(age**2) ', mydata1).fit()
model3.summary()
```

The estimated effect of education here is -0.1559, which is less than -0.0906 in part a).
