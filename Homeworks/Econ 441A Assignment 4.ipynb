```python
import patsy as pt
import statsmodels.stats.outliers_influence as smo

y,X = pt.dmatrices('np.log(psoda)~prpblck+np.log(income)+prppov', data = mydata,
                   return_type = 'dataframe')

X.head()
```

```python
k = X.shape[1]

VIF = np.empty(k)

for i in range(k):
    
    VIF[i] = smo.variance_inflation_factor(X.values, i)

print('VIF:', VIF)
```

```python
reg2=smf.ols('np.log(psoda)~prpblck+np.log(income)+prppov+np.log(hseval)',data=mydata)
results2=reg2.fit()
results2.summary()
```

The coefficient is 0.1213, which means 1% increase in the median housing value will increase the price of medium soda by 12.13%. And the p-value is 0.000, which is less than 0.005, it means that we need to reject H0 and the coefficient of log(hseval) is statistically significant.

### 1/d)

```python
hypotheses = ["np.log(income)=0","prppov=0"]
results2.f_test(hypotheses)
```

In the new regression model above, the p-value for log(income) is 0.159 and p-value for prppov is 0.699. Since 0.159>0.05, 0.699>0.05, log(income) and prppov are not statistically significant as individual variables.   
Since the p-value for the F-Test is 0.0304, which is less than 0.05, we reject the null that both coefficients of log(income) and prppov are equal to zero. So these two variables are jointly significant.

```python
reg3=smf.ols("voteA~prtystrA+np.log(expendA)+democA+np.log(expendB)",data=mydata1)
results3=reg3.fit()
Residual=results3.resid
Residual
results3.summary()
```

```python
reg4=smf.ols("Residual~prtystrA+np.log(expendA)+democA+np.log(expendB)",data=mydata1)
results4=reg4.fit()
results4.summary()
```

R^2=SSE/SST=Σ(ŷ-ȳ)^2/SST.Since the sum of all residuals equal to 0, ȳ=0.For ŷ, ŷ=y-residual(y), since y is the residual from last model which does not explain the variation of last model at all, the residual of y is the same as y which does not explain the model at all. Then ŷ=0, so SSE=0, so R^2=0.

### 2/b)

```python
mydata1["res2"] = Residual**2

aux_reg = smf.ols('res2~prtystrA+np.log(expendA)+democA+np.log(expendB)', mydata1).fit()

f = aux_reg.fvalue
fp = aux_reg.f_pvalue

print("The F-Statistic for the Auxiliary Regression is: "+ str(f) +" and the P-Value is: "+ str(fp))
```

Since the p-value is greater than 0.05, we can conclude that heteroscedasticity is not present in the model.

```python
reg7 = smf.ols('voteA~prtystrA+np.log(expendA)+democA+np.log(expendB)+shareA+district', mydata1)
reg7.fit().summary()
```

```python
mydata1["ehatsq"] = reg7.fit().resid**2

# estimate weights
w_est = smf.ols('np.log(ehatsq) ~ prtystrA+np.log(expendA)+democA+np.log(expendB)+shareA+district', data = mydata1).fit()

vari = np.exp(w_est.fittedvalues) #estimated variances
w = 1/vari**2

fgls =smf.wls('voteA~ prtystrA+np.log(expendA)+democA+np.log(expendB)+shareA+district', mydata1, weights = w).fit()

fgls.summary()
```

The coefficients, standard errors, and p-values do not change a lot after using the FGLS.
